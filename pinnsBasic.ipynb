```python
# gravity thing
g = 9.81
dt = 0.1
t_max = 2.0     # 2 secs

v, y, t = 0, 0, 0   # initial velocity, position, time
print("start of the velocity", v)
while t < t_max:
    v = v + g * dt
    y = y + v * dt
    t = t + dt
    print(f"t={t:.1f}s, y={y:.2f}m, v={v:.2f}m/s")

print('final velo:',v)
```

```python
# for a 2d projectile motion,starting from 0,0 and with no air registance we have g.
import math
import matplotlib.pyplot as plt
```

```python
g = 9.81
dt = 0.1
vx = 10 * math.cos(math.radians(45))  # initial velocity components
vy = 10 * math.sin(math.radians(45))
x, y, t = 0, 0, 0
X, Y = [], []


while y >= 0:
    vy = vy - g * dt
    x  = x + vx * dt
    y  = y + vy * dt
    t  = t + dt
    print(f"t={t:.1f}s, x={x}m, y={y}m")
    X.append(x)
    Y.append(y)

plt.plot(X, Y)
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.show()
```
